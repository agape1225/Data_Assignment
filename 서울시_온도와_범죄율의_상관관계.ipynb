```python
import pandas as pd
import csv

seoul_crime_statistics = pd.read_csv('201904055.csv', encoding='utf-8-sig', low_memory=False, quoting=csv.QUOTE_NONE)

#서울시의 온도와 범죄율 데이터
seoul_crime_statistics.head(20)
```

데이터를 비교해야 해서 하나의 피겨에 여러개의 선을 그려야합니다.

```python
import matplotlib.pyplot as plt

col_list = seoul_crime_statistics["일시"].values.tolist()
tem = seoul_crime_statistics["평균기온(℃)"].values.tolist()
print(col_list)
print(tem)
```

```python
fig, ax = plt.subplots(figsize=(60,20))
ax.plot(col_list, tem)
ax.set_title("seoul temp")
ax.legend()
fig.show()
```

```python
import re

crime_sum = seoul_crime_statistics["합계 발생"].values.tolist()
crime_sum = [re.sub(r'[^0-9]', '', str(elem)) for elem in crime_sum]
crime_sum = list(map(int, crime_sum))

murder = seoul_crime_statistics["살인 발생"].values.tolist()
murder = [re.sub(r'[^0-9]', '', str(elem)) for elem in murder]
murder = list(map(int, murder))

robbery = seoul_crime_statistics["강도 발생"].values.tolist()
robbery = [re.sub(r'[^0-9]', '', str(elem)) for elem in robbery]
robbery = list(map(int, robbery))

rape = seoul_crime_statistics["강간.강제추행 발생"].values.tolist()
rape = [re.sub(r'[^0-9]', '', str(elem)) for elem in rape]
rape = list(map(int, rape))

theft = seoul_crime_statistics["절도 발생"].values.tolist()
theft = [re.sub(r'[^0-9]', '', str(elem)) for elem in theft]
theft = list(map(int, theft))

violence = seoul_crime_statistics["폭력 발생"].values.tolist()
violence = [re.sub(r'[^0-9]', '', str(elem)) for elem in violence]
violence = list(map(int, violence))
```

```python
fig, ax = plt.subplots(figsize=(60,20))
#ax.plot(col_list, crime_sum, label="crime_sum")
ax.plot(col_list, murder, label="murder")
ax.plot(col_list, robbery, label="robbery")
ax.plot(col_list, rape, label="rape")
ax.plot(col_list, theft, label="theft")
ax.plot(col_list, violence, label="violence")

ax.legend()

ax.set_title("seoul crime rate")
fig.show()
```

이렇게 해보니 온도 데이터와는 차이가 너무 나서 정규화 과정을 거치게 되었습니다.

```python
normaliz_data = seoul_crime_statistics
#필요없는 데이터는 가시성을 위해 삭제하였습니다.

normaliz_data = normaliz_data.drop(["평균최고기온(℃)",	"최고기온(℃)",	
                    "최고기온일자",	"평균최저기온(℃)",	
                    "최저기온(℃)",	"최저기온일자",	"합계 발생"], axis="columns")
normaliz_data.head()
```

```python
def apply_formula(value):
    value = re.sub(r'[^0-9]', '', str(value))
    value = int(value)
    return value

normaliz_data.iloc[:, 2:] = normaliz_data.iloc[:, 2:].applymap(apply_formula)

normaliz_data.head()
```

```python
normaliz_data_index = normaliz_data.set_index("일시")
normaliz_data_index.head()
```

```python
normaliz_data_trans = normaliz_data_index.transpose()
#normaliz_data_trans.head()
date_col = normaliz_data_trans.columns.get_level_values(0)
print(date_col)
```

```python
fig, ax = plt.subplots(figsize=(60,20))

ax.stackplot(date_col, normaliz_data_trans, labels=[	"avr_temp",	"murder",	"robbery",	"rape",	"theft","violence"])
ax.legend(loc='upper left')

fig.show()
```

```python
normaliz_data.head()

# Min-Max 정규화 함수
def normalize(column):
    return (column - column.min()) / (column.max() - column.min())

# 평균기온 열을 제외한 모든 열에 정규화 적용
final_data = normaliz_data.iloc[:, 1:] = normaliz_data.iloc[:, 1:].apply(normalize)

# 정규화된 데이터프레임 출력
final_data.head()
```

```python
fig, ax = plt.subplots(figsize=(60,20))
#ax.plot(col_list, crime_sum, label="crime_sum")
ax.plot(col_list, final_data["평균기온(℃)"], label="temp")
ax.plot(col_list, final_data["살인 발생"], label="murder")
ax.plot(col_list, final_data["강도 발생"], label="robbery")
ax.plot(col_list, final_data["강간.강제추행 발생"], label="rape")
ax.plot(col_list, final_data["절도 발생"], label="theft")
ax.plot(col_list, final_data["폭력 발생"], label="violence")

ax.legend()

ax.set_title("seoul crime rate")
fig.show()
```
