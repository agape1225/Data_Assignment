# 데이터 합치기

서울시의 범죄 데이터는 행과 열이 바뀐 상태였습니다. 따라서 trsanspose 함수를 활용해 해당 데이터 프레임의 행과 열을 바꿔주었습니다.

바꿔준 데이터를 날짜별로 정리하고 인덱스를 초기화하여 해당 인덱스를 기준으로 INNER JOIN을 실행해 최종적으로 데이터를 합쳐주었습니다.

```python
seoul_weather["발생 합계"] = 0

seoul_weather.head()

seoul_crime_columns = seoul_crime.columns

seoul_crime=seoul_crime.transpose()

print(seoul_crime.head())

selected_columns = [True if seoul_crime[buff][2] == '"발생"' else False for buff in seoul_crime.columns]
seoul_crime = seoul_crime.loc[:,selected_columns]

#print(selected_columns)

seoul_crime.columns = ["합계 발생", "살인 발생", "강도 발생", "강간.강제추행 발생", "절도 발생", "폭력 발생"]
seoul_crime = seoul_crime[3:]
seoul_crime.head()
```
